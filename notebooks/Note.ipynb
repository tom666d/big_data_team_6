### Direction 2: Multi-Agent Architecture Upgrade
**Current architecture:**
Spark → Diagnosis Agent → Remediation Agent → Human Review

**Proposed upgrade:**
```
Spark Detection
    ↓
Router Agent (classifies issue type)
    ↓
Null Spike    → Null Specialist Agent
Outlier       → Outlier Specialist Agent  
Format Error  → Format Specialist Agent
    ↓
Critic Agent (risk assessment)
"Is this fix safe? Any data loss risk?"
    ↓
Human Review (Streamlit)
```
**Why it matters:**
- Router = "triage system" analogy → instantly understandable
- Critic = three-layer AI governance → strong differentiator
- Industry-standard Multi-Agent pattern (LangGraph, CrewAI)

**Pitch to evaluators:**
> "Our system has four layers of AI protection:
> Spark detection → Router triage → Specialist remediation
> → Critic risk assessment → Human approval.
> Every fix is validated by AI before a human ever sees it."

**Estimated effort:** 3-4 days (Week 3)
